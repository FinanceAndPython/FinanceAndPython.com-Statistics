```python
xVals = list(range(100))
yVals = [x+randint(-20,20) for x in xVals]
xVals = np.array(xVals)
yVals = np.array(yVals)
print(yVals-yVals.mean())
```

```python
print((yVals-yVals.mean())/yVals.std())
```

```python
z1 = (xVals-xVals.mean())/xVals.std()
z2 = (yVals-yVals.mean())/yVals.std()
print(sum(z1*z2)/(100))
print(np.corrcoef(xVals,yVals)[0][1])
```

```python
import scipy.stats
result = scipy.stats.linregress(xVals,yVals)
print(result)
```

```python
print(result.slope)
print(result.intercept)
```

```python
print(result.rvalue)
print(result.pvalue)
```

```python
m = result.slope
b = result.intercept
fittedValues = [m*x+b for x in xVals]
print(fittedValues)
```

```python
plt.scatter(xVals,yVals)
plt.plot(xVals,fittedValues,"r")
plt.show()
```

```python
yVals = np.array(yVals)
fittedValues = np.array(fittedValues)
residuals = yVals-fittedValues
print(residuals)
plt.scatter(xVals,residuals)
plt.show()
```

```python
#heteroskedastic results! We want the residuals to not look like they have a trend
```

```python
xVals = list(range(100))
yVals = [x**2 for x in xVals]

result = scipy.stats.linregress(xVals,yVals)
m = result.slope
b = result.intercept
fittedValues = [m*x+b for x in xVals]

plt.scatter(xVals,yVals)
plt.plot(xVals,fittedValues,"r")
plt.title("Regression")
plt.show()
yVals = np.array(yVals)
fittedValues = np.array(fittedValues)
residuals = yVals-fittedValues
plt.scatter(xVals,residuals)
plt.title("Residuals")
plt.show()
```

```python
xVals = list(range(100))
yVals = [x+randint(-5, 5)*x/5for x in xVals]

result = scipy.stats.linregress(xVals,yVals)
m = result.slope
b = result.intercept
fittedValues = [m*x+b for x in xVals]

plt.scatter(xVals,yVals)
plt.plot(xVals,fittedValues,"r")
plt.title("Regression")
plt.show()
yVals = np.array(yVals)
fittedValues = np.array(fittedValues)
residuals = yVals-fittedValues
plt.scatter(xVals,residuals)
plt.title("Residuals")
plt.show()
```
