```python
import scipy.stats
dist = scipy.stats.norm(400,100)
pts = dist.rvs(100,random_state=1)
print(pts)
```

```python
import numpy as np
mean = np.mean(pts)
std = np.std(pts)
t = (mean-400)/(std/len(pts)**.5)
print(t)
```

```python
SE = std/(len(pts)**.5)
bounds = (mean-SE*t,mean+SE*t)
print(bounds)
```

```python
(.5-scipy.stats.norm.cdf(-t))*2
```

```python
leftX = np.arange(-4,-1.95,0.01)
middleX = np.arange(-1.96,1.97,0.01)
rightX = np.arange(1.96,4.01,0.01)
print(leftX)
print(middleX)
print(rightX)
```

```python
import matplotlib.pyplot as plt
leftY = scipy.stats.norm.pdf(leftX)
middleY = scipy.stats.norm.pdf(middleX)
rightY = scipy.stats.norm.pdf(rightX)
plt.rcParams["figure.figsize"] = [10,10]
plt.plot(leftX,leftY,"r")
plt.plot(middleX,middleY,"gray")
plt.plot(rightX,rightY,"r")
plt.xlabel("Z Score")
plt.ylabel("PDF")
plt.title("Two Tail 95% Confidence Interval")
plt.fill_between(leftX,leftY,facecolor="r")
plt.fill_between(rightX,rightY,facecolor="r")
plt.fill_between(middleX,middleY,facecolor="gray")
plt.text(-2.25,.01,"2.5%",horizontalalignment='center')
plt.text(2.25,.01,"2.5%",horizontalalignment='center')
plt.text(0,.01,"95%",horizontalalignment='center')

plt.show()
```
